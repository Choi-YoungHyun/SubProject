# 3. 패스트푸드 매장(롯데리아, 맥도날드, 버거킹, 맘스터치)은 어느 지역에 많이 분포할까? (시도)

* 패스트푸드 매장의 지역 분포를 알아봐야한다.

* 패스트푸드에 해당하는 매장만 df_Fast로 분리하고 인덱스를 재정렬한다.

* 그 중 '시도명'이라는 column을 가지고 지역별 개수 파악이 가능하다

* df_Fast['시도명'].value.counts()로 개수 파악을 했을 때 경기도가 제일 많았다.

* 이해를 돕기 위해 시각화 그래프를 그렸다. (barplot으로과 scatterplot)

* 결론: 경기도가 압도적으로 개수가 많았고, 2위로 서울 3위로 경상남도, 4위로 부산 등 순위가 이어졌다.

* scatterplot으로 보았을 때 수도권이 압도적으로 많고 또한 주로 광역시 및 대도시에 패스트푸드 매장이 집중적으로 있다는 점을 알수 있다.

* 면적 대비 작은 면적을 가진 제주도에도 상당한 점포가 있는 것도 알 수 있다.

```python
df_Fast = df_total.loc[(df_total['상호명'].str.contains('맥도날드')) 
                        |(df_total['상호명'].str.contains('롯데리아')) 
                        |(df_total['상호명'].str.contains('버거킹')) 
                        |(df_total['상호명'].str.contains('맘스터치'))].reset_index()


df_Fast_Map = pd.DataFrame(df_Fast['시도명'].value_counts())
df_Fast_Map.columns=['지역별 개수']

plt.figure(figsize=(20,12))

plt.bar(df_Fast_Map.index,df_Fast_Map['지역별 개수'],alpha=0.5,color='r')

plt.show()
```
