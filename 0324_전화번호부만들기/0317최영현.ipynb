```python
def intro():
    print('\n아래번호를 선택하세요\n1.전화번호부 입력\n2.전화번호부 출력\n3.전화번호부 저장\n4.전화번호부 검색\n5.전화번호부 삭제\n6.전화번호부 종료')
    global choice
    choice = int(input())
    # 반복되는 print를 함수로 저장
    # choice는 꼭 int 값이여야되서 글로벌 지정

def search_program(word):
    file = open ('./0317최영현','rt')
    before_list = file.read()
    global temp_list
    temp_list = before_list.split('\n') # 결과는 리스트(가로출력)
    # 재실행 했을 시 저장되있는 데이터를 불러오기 위해 
    
    for i in range(len(temp_list)):
        if word in temp_list[i]:
            print(temp_list[i]) #결과 list
    file = open('./0317최영현','at')
    file.close()
    # 리스트 안에 str값이 들어가있는지 in으로 찾은 후 True면 출력

def before():
    file = open ('./0317최영현','rt')
    before_list = file.read()
    temp2_list = before_list.split('\n')
    return temp2_list
    #search를 먼저 만들다가 조건을 추가하다보니 
    #파일에 저장된 데이터를 가져오는 일이 많아 작성 
```

```python
try:
    file = open('./0317최영현','at')
except:
    print('에러')

bool1= True
total_list = [] #메모리에 올린 데이터를 넣는 곳 
intro()

while bool1 :
    if choice == 1:
        try:
            name,hp,address,email,group = input('이름,번호,주소,이메일,그룹을 컴마 기준으로 나누어 쓰시오.\n띄어쓰기불가 : ').split(',')
            # for와 if 문으로 중복된 값 찾는 검사를 시행
            # 만들어놓은 before() = 저장된 값을 \n으로 스플릿한 리스트 
            for i in range(len(before())):
                if hp in before()[i]:
                    print('중복된 전화번호입니다.\n 초기화면으로 돌아갑니다.')
                    intro()
            
            total = (f'이름:{name},전화번호:{hp},주소{address},이메일:{email},소속:{group}')
            total_list.append(total)
                    #오름차순 정렬 
                    #sorted_list에 sorted함수를 적용 후 
                    #x 값이 계속 변해야 되는데 for문을 쓸까하다가, 람다 적용)
                    #sort는 제가 원하는 기준을 적고 싶지만 parameter가 없어서 사용x
            sorted_list = sorted(total_list,key = lambda x:x.split(',')[0])
            total_list = sorted_list
                
            print('\n입력되었습니다.\n✔저장을 먼저해주시길 바랍니다.')
        except ValueError:
            print('\n값이 부족하거나, 너무 많습니다.\n초기화면으로 이동')
            # 5개의 변수를 한번의 input으로 받아서, 값이 부족하거나, 많을 수 있어서 작성
        finally:
            intro()
        

    elif choice == 2:
        file = open('./0317최영현','r')
        print('기존 데이터')
        data = file.read()
        print(data)

        print('새로 추가한 데이터')
        for i in range(len(total_list)):
            print(total_list[i])

        file = open('./0317최영현','at')

        intro()
    
    elif choice == 3:
        for j in range(len(total_list)):
            file.write(str(total_list[j]) + '\n')
            #리스트 형식으로 저장되지도 않고, 엔터키로 구분해야되서 str으로 변경
        sum_word = len(total_list)
        print(f'{sum_word}개가 저장되었습니다.')
        intro()

    elif choice == 4:
            print('검색할 데이터 1개를 입력하시오.\n예시) 홍길동 혹은 010-2222-2222')
            choice2 = input()
            search_program(choice2)
            #만든 search_program 사용

            intro()

    elif choice == 5:
        del1 = input('\n삭제할 데이터를 입력하세요.\n✔단 저장된 데이터만 삭제 가능합니다.')

        temp_list = before()
        #만든 before 함수 적용

        # 혹시 삭제 된 리스트가 존재 할까봐 -1으로 시작 
        # 그렇지 않으면 IndexError:  out range of list 떠서 사용
        for i in range(len(temp_list)-1):
            if del1 in temp_list[i]:
                temp_list.remove(temp_list[i])
                file = open('./0317최영현','wt')
                file.write(str(temp_list)+'\n')
                file.close()
        print('삭제되었습니다')
        intro()        

    else:
        print('프로그램을 종료합니다.')
        file.close()
        bool1= False
```
